```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 读取 CSV 文件
file_path = "exchange_rates_2015_2024.csv"
df = pd.read_csv(file_path)

# 重命名汇率列
df.rename(columns={"DEXSZUS": "chf", "DEXCHUS": "cny"}, inplace=True)

# 转换日期列为 datetime 类型，并按时间排序
df["observation_date"] = pd.to_datetime(df["observation_date"])
df.sort_values("observation_date", inplace=True)

# 对 chf 和 cny 取对数（加一个检查防止0或负数）
df["log_chf"] = np.log(df["chf"])
df["log_cny"] = np.log(df["cny"])

# 绘制时间序列图
plt.figure(figsize=(12, 6))
plt.plot(df["observation_date"], df["log_chf"], label="log(chf)")
plt.plot(df["observation_date"], df["log_cny"], label="log(cny)")
plt.xlabel("Date")
plt.ylabel("Log Exchange Rate")
plt.title("Log of Exchange Rates Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 划分训练集和测试集（按时间顺序）
split_index = int(len(df) * 0.8)
train_df = df.iloc[:split_index]
test_df = df.iloc[split_index:]

# 打印划分结果
print("训练集大小:", len(train_df))
print("测试集大小:", len(test_df))
print("训练集起止日期:", train_df["observation_date"].min(), "→", train_df["observation_date"].max())
print("测试集起止日期:", test_df["observation_date"].min(), "→", test_df["observation_date"].max())
```

```python
print("CHF 差分的均值：", train_df['diff_chf'].mean())
print("CNY 差分的均值：", train_df['diff_cny'].mean())
```

```python
# 先对数化，再乘以100
train_df["log_chf_100"] = np.log(train_df["chf"]) * 100
train_df["log_cny_100"] = np.log(train_df["cny"]) * 100

# 差分
train_df["diff_log_chf"] = train_df["log_chf_100"].diff()
train_df["diff_log_cny"] = train_df["log_cny_100"].diff()

# 均值（趋势）
a_chf = train_df["diff_log_chf"].mean()
a_cny = train_df["diff_log_cny"].mean()

# 减去均值，得残差序列
train_df["resid_log_chf"] = train_df["diff_log_chf"] - a_chf
train_df["resid_log_cny"] = train_df["diff_log_cny"] - a_cny

# 打印趋势值
print(f"趋势 a_chf: {a_chf:.6f}")
print(f"趋势 a_cny: {a_cny:.6f}")

# 绘图
plt.figure(figsize=(12, 6))
plt.plot(train_df.index, train_df["resid_log_chf"], label="resid_log_chf = Δlog(chf×100) - a")
plt.plot(train_df.index, train_df["resid_log_cny"], label="resid_log_cny = Δlog(cny×100) - a")
plt.axhline(0, color="gray", linestyle="--")
plt.xlabel("Date")
plt.ylabel("Residual (Δlog * 100 - a)")
plt.title("Residuals of Log-Differenced CHF & CNY (×100)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import het_arch, acorr_ljungbox
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# 准备两个序列
resid_cny = train_df['resid_cny'].dropna()
resid_chf = train_df['resid_chf'].dropna()

# 包装成一个函数重复使用
def check_garch_applicability(series, name='Series'):
    print(f"\n==== {name} 检验结果 ====")
    
    # ADF 单位根检验
    adf_result = adfuller(series)
    print(f"{name} ADF 检验 p值（平稳性）: {adf_result[1]:.4f}")
    
    # ARCH 效应检验
    arch_test = het_arch(series)
    print(f"{name} ARCH 效应检验 p值: {arch_test[1]:.4f}")
    
    # Ljung-Box 检验（对平方残差）
    squared = series**2
    lb_test = acorr_ljungbox(squared, lags=[10], return_df=True)
    print(f"{name} Ljung-Box 检验在平方残差上的 p值（lags=10）: {lb_test['lb_pvalue'].values[0]:.4f}")
    
    # 绘图
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    axes[0, 0].plot(series)
    axes[0, 0].set_title(f"{name} Residual Series")
    
    axes[0, 1].plot(squared)
    axes[0, 1].set_title(f"{name} Squared Residuals")
    
    stats.probplot(series, dist="norm", plot=axes[1, 0])
    axes[1, 0].set_title(f"{name} QQ Plot")
    
    sns.histplot(series, bins=50, kde=True, ax=axes[1, 1])
    axes[1, 1].set_title(f"{name} Histogram")
    
    plt.tight_layout()
    plt.show()

# 对 resid_cny 和 resid_chf 分别检查
check_garch_applicability(resid_cny, name="CNY")
check_garch_applicability(resid_chf, name="CHF")
```

```python
import numpy as np

def simulate_garch_paths(fitted_model, T=30, n_paths=1000, seed=42):
   
    np.random.seed(seed)
    params = fitted_model.params
    omega = params["omega"]
    alpha = params["alpha[1]"]
    beta = params["beta[1]"]

    # 初始化数组
    epsilons = np.zeros((n_paths, T))
    sigmasq = np.zeros((n_paths, T))

    # 初始值：使用样本的最后条件方差
    last_vol = fitted_model.conditional_volatility.iloc[-1]
    last_eps = fitted_model.std_resid.iloc[-1]

    for i in range(n_paths):
        sigma2 = last_vol**2
        eps = last_eps

        for t in range(T):
            z = np.random.normal()
            sigma2 = omega + alpha * eps**2 + beta * sigma2
            sigma = np.sqrt(sigma2)
            eps = z * sigma

            epsilons[i, t] = eps
            sigmasq[i, t] = sigma2

    return epsilons.mean(axis=0), epsilons  # 返回平均路径和全部模拟序列

# 模拟 30天、180天、365天路径
durations = [30, 180, 365]

# 保存结果
sim_results = {}

for duration in durations:
    mean_path_cny, _ = simulate_garch_paths(res_cny, T=duration, n_paths=1000)
    mean_path_chf, _ = simulate_garch_paths(res_chf, T=duration, n_paths=1000)
    sim_results[duration] = {
        "CNY": mean_path_cny,
        "CHF": mean_path_chf
    }

# 打印结果（例如前5步）
for duration in durations:
    print(f"\n=== {duration} 天平均残差路径 ===")
    print("CNY 前5项:", np.round(sim_results[duration]['CNY'][:5], 4))
    print("CHF 前5项:", np.round(sim_results[duration]['CHF'][:5], 4))
```

```python
cum_sim_cny_30d = np.cumsum(sim_cny_30d)
cum_sim_cny_180d = np.cumsum(sim_cny_180d)
cum_sim_cny_365d = np.cumsum(sim_cny_365d)

# 打印前5项 + 总长度
print(" CNY 模拟累计路径")
print("30天:", np.round(cum_sim_cny_30d[:5], 6), "...", f"(共 {len(cum_sim_cny_30d)} 项)")
print("180天:", np.round(cum_sim_cny_180d[:5], 6), "...", f"(共 {len(cum_sim_cny_180d)} 项)")
print("365天:", np.round(cum_sim_cny_365d[:5], 6), "...", f"(共 {len(cum_sim_cny_365d)} 项)")
```

```python
# 对 CHF 模拟路径进行逐步累加
cum_sim_chf_30d = np.cumsum(sim_chf_30d)
cum_sim_chf_180d = np.cumsum(sim_chf_180d)
cum_sim_chf_365d = np.cumsum(sim_chf_365d)

# 打印前5项 + 总长度
print(" CHF 模拟累计路径")
print("30天:", np.round(cum_sim_chf_30d[:5], 6), "...", f"(共 {len(cum_sim_chf_30d)} 项)")
print("180天:", np.round(cum_sim_chf_180d[:5], 6), "...", f"(共 {len(cum_sim_chf_180d)} 项)")
print("365天:", np.round(cum_sim_chf_365d[:5], 6), "...", f"(共 {len(cum_sim_chf_365d)} 项)")
```

```python
# 获取训练集最后的 log 汇率 *100 值
last_log_chf = train_df["log_chf_100"].iloc[-1]
last_log_cny = train_df["log_cny_100"].iloc[-1]

# 获取趋势项均值 a
a_chf = train_df["diff_log_chf"].mean()
a_cny = train_df["diff_log_cny"].mean()

# 构造 t = [1, 2, ..., T]
t_30 = np.arange(1, 31)
t_180 = np.arange(1, 181)
t_365 = np.arange(1, 366)

# 最终预测值 = last + cum_residual + t*a
forecast_chf_30d = (last_log_chf + cum_sim_chf_30d + a_chf * t_30)/100
forecast_chf_180d = (last_log_chf + cum_sim_chf_180d + a_chf * t_180)/100
forecast_chf_365d = (last_log_chf + cum_sim_chf_365d + a_chf * t_365)/100

forecast_cny_30d = (last_log_cny + cum_sim_cny_30d + a_cny * t_30)/100
forecast_cny_180d = (last_log_cny + cum_sim_cny_180d + a_cny * t_180)/100
forecast_cny_365d = (last_log_cny + cum_sim_cny_365d + a_cny * t_365)/100

```
