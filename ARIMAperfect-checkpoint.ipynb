```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller, kpss

# 读取数据
df = pd.read_csv("exchange_rates_2015_2024.csv")

# 重命名列
df = df.rename(columns={'DEXSZUS': 'chf', 'DEXCHUS': 'cny'})

# 对 chf 和 cny 取对数
df['log_chf'] = np.log(df['chf'])
df['log_cny'] = np.log(df['cny'])

# ADF 和 KPSS 检验函数
def print_stationarity_tests(series, name):
    adf_result = adfuller(series.dropna())
    kpss_result = kpss(series.dropna(), regression='c', nlags="auto")
    print(f"--- {name} ---")
    print(f"ADF Statistic: {adf_result[0]:.4f}, p-value: {adf_result[1]:.4f}")
    print(f"KPSS Statistic: {kpss_result[0]:.4f}, p-value: {kpss_result[1]:.4f}")
    print()

# 打印两个变量的平稳性检验
print_stationarity_tests(df['log_chf'], 'log_chf')
print_stationarity_tests(df['log_cny'], 'log_cny')

# 划分训练集（前80%）和测试集（后20%）
split_index = int(len(df) * 0.8)
train = df.iloc[:split_index].copy()
test = df.iloc[split_index:].copy()

# 对 log_chf 和 log_cny 乘以 100（仅训练集）
train['log_chf_100'] = train['log_chf'] * 100
train['log_cny_100'] = train['log_cny'] * 100

# 差分
log_chf_diff = train['log_chf_100'].diff().dropna()
log_cny_diff = train['log_cny_100'].diff().dropna()

# 均值（趋势项）
trend_chf = log_chf_diff.mean()
trend_cny = log_cny_diff.mean()
print(f"趋势 a_chf: {trend_chf:.6f}")
print(f"趋势 a_cny: {trend_cny:.6f}")

# 去趋势处理
detrended_log_chf = log_chf_diff - trend_chf
detrended_log_cny = log_cny_diff - trend_cny

# 打印前几行去趋势后的结果
print("\n去趋势后的 log_chf 差分序列：")
print(detrended_log_chf.head())

print("\n去趋势后的 log_cny 差分序列：")
print(detrended_log_cny.head())
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Rename detrended series
et_chf = detrended_log_chf
et_cny = detrended_log_cny

# 1. Time series plots
plt.figure(figsize=(14, 5))
plt.plot(et_chf, label='et_chf')
plt.title("Time Series of Detrended Residuals (et_chf)")
plt.xlabel("Time")
plt.ylabel("Residual Value")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 5))
plt.plot(et_cny, label='et_cny', color='orange')
plt.title("Time Series of Detrended Residuals (et_cny)")
plt.xlabel("Time")
plt.ylabel("Residual Value")
plt.legend()
plt.grid(True)
plt.show()

# 2. ADF and KPSS test
print_stationarity_tests(et_chf, 'et_chf')
print_stationarity_tests(et_cny, 'et_cny')

# 3. ACF and PACF plots
fig, ax = plt.subplots(2, 1, figsize=(12, 6))
plot_acf(et_chf, ax=ax[0], lags=30)
ax[0].set_title("ACF of et_chf")
plot_pacf(et_chf, ax=ax[1], lags=30, method='ywm')
ax[1].set_title("PACF of et_chf")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(2, 1, figsize=(12, 6))
plot_acf(et_cny, ax=ax[0], lags=30)
ax[0].set_title("ACF of et_cny")
plot_pacf(et_cny, ax=ax[1], lags=30, method='ywm')
ax[1].set_title("PACF of et_cny")
plt.tight_layout()
plt.show()
```

```python
from pmdarima import auto_arima

# 设置自动 ARMA 拟合函数（排除 (0,1,0)）
def fit_auto_arma(series, name):
    print(f"\n--- Auto ARMA for {name} ---")
    model = auto_arima(series,
                       start_p=0, start_q=0,
                       max_p=5, max_q=5,
                       d=0,                      # 不进行差分
                       seasonal=False,
                       stepwise=True,
                       suppress_warnings=True,
                       error_action='ignore',
                       max_order=10,
                       information_criterion='aic',
                       n_fits=30)
    
    # 手动排除 (0,1,0) 结果（如果出现）
    if model.order == (0, 1, 0):
        print(f"Model selected (0,1,0), retrying with restriction...")
        model = auto_arima(series,
                           start_p=0, start_q=0,
                           max_p=5, max_q=5,
                           d=0,
                           seasonal=False,
                           stepwise=True,
                           suppress_warnings=True,
                           error_action='ignore',
                           max_order=10,
                           information_criterion='aic',
                           n_fits=30,
                           allow_order=(lambda p, d, q: not (p == 0 and d == 1 and q == 0)))
    
    print(f"Selected ARMA order: {model.order}")
    print(f"AIC: {model.aic():.4f}")
    print(f"BIC: {model.bic():.4f}")
    print(model.summary())

# 对两个序列执行
fit_auto_arma(et_chf, 'et_chf')
fit_auto_arma(et_cny, 'et_cny')
```

```python
import warnings
from statsmodels.tsa.arima.model import ARIMA
warnings.filterwarnings("ignore")
# 要拟合的模型阶数组（p,d,q）
model_orders = [(0,0,1), (1,0,1), (2,0,1), (1,0,2), (0,0,2), (2,0,0)]

def evaluate_arma_models(series, name):
    print(f"\n=== Evaluating ARMA Models for {name} ===")
    for order in model_orders:
        try:
            model = ARIMA(series, order=order)
            result = model.fit()
            print(f"ARMA{order} - AIC: {result.aic:.4f}, BIC: {result.bic:.4f}")
        except Exception as e:
            print(f"ARMA{order} - Failed to fit: {e}")

# 分别对 et_chf 和 et_cny 进行拟合
evaluate_arma_models(et_chf, 'et_chf')
evaluate_arma_models(et_cny, 'et_cny')
```

```python
import numpy as np

# 模型系数
phi1_chf, phi2_chf, theta1_chf = 1.0904, -0.1488, -0.9754
phi1_cny, phi2_cny, theta1_cny = 0.9212, 0.0430, -0.9437

# 残差标准差
a_chf = 0.5742
a_cny = 0.2382

# 从训练集中获取最后两个真实 et 值（你应提前定义 et_chf 和 et_cny）
et_chf_last2 = et_chf.values[-2:]  # [et[-2], et[-1]]
et_cny_last2 = et_cny.values[-2:]

# 初始 ε[-1] 设为 0
eps0 = 0.0

# 模拟参数
n_sim = 1000
horizons = [30, 180, 365]

# 模拟函数：从真实 et[-2], et[-1] 开始，eps[-1] = 0
def simulate_arma_given_init(phi1, phi2, theta1, std, init_et, steps, n_sim=1000):
    sims = np.zeros((n_sim, steps))
    for i in range(n_sim):
        et = np.zeros(steps + 2)
        eps = np.random.normal(0, std, size=steps + 2)
        # 初始条件
        et[0], et[1] = init_et[0], init_et[1]
        eps[1] = eps0  # 初始 epsilon[t-1] = 0
        for t in range(2, steps + 2):
            et[t] = phi1 * et[t - 1] + phi2 * et[t - 2] + theta1 * eps[t - 1] + eps[t]
        sims[i] = et[2:]
    return sims

# 执行模拟
results_chf = {h: simulate_arma_given_init(phi1_chf, phi2_chf, theta1_chf, a_chf, et_chf_last2, h, n_sim) for h in horizons}
results_cny = {h: simulate_arma_given_init(phi1_cny, phi2_cny, theta1_cny, a_cny, et_cny_last2, h, n_sim) for h in horizons}

# 计算平均路径
mean_path_chf = {h: np.mean(results_chf[h], axis=0) for h in horizons}
mean_path_cny = {h: np.mean(results_cny[h], axis=0) for h in horizons}
```

```python
# 重新构造 log 差分序列
log_chf_diff_recovered = detrended_log_chf + trend_chf
log_cny_diff_recovered = detrended_log_cny + trend_cny

# 从差分恢复原始 log 值（乘以100）序列
# 提供初始值：log_chf_100[0] 和 log_cny_100[0]
log_chf_100_recovered = log_chf_diff_recovered.cumsum() + train['log_chf_100'].iloc[1]
log_cny_100_recovered = log_cny_diff_recovered.cumsum() + train['log_cny_100'].iloc[1]

# 取最后一个 wt 值（你要的起点 w0）
w0_chf = log_chf_100_recovered.iloc[-1]
w0_cny = log_cny_100_recovered.iloc[-1]

print("w0_chf =", round(w0_chf, 6))
print("w0_cny =", round(w0_cny, 6))
```

```python

import numpy as np

# 初始值（训练集最后一个 wt）
w0_chf = -3.71391
w0_cny = 190.902385

# 趋势项（从 log 差分序列中获得）
trend_chf = log_chf_diff.mean()
trend_cny = log_cny_diff.mean()

# === 一步一步还原 ===

# 累加 et + w0 → 还原 wt（无趋势）
def recover_w_from_et_only(mean_et, w0):
    return np.cumsum(mean_et) + w0

# 加上线性趋势（t * a）
def add_trend_to_wt(wt_series, trend):
    return wt_series + np.arange(len(wt_series)) * trend

# 还原 log 汇率路径（wt / 100）
def restore_log_rate(mean_et, w0, trend):
    wt = recover_w_from_et_only(mean_et, w0)
    wt_trended = add_trend_to_wt(wt, trend)
    return wt_trended / 100  # log(rate)

# 最终 log 汇率预测路径（无需 exp）
log_chf_30 = restore_log_rate(mean_path_chf[30], w0_chf, trend_chf)
log_chf_180 = restore_log_rate(mean_path_chf[180], w0_chf, trend_chf)
log_chf_365 = restore_log_rate(mean_path_chf[365], w0_chf, trend_chf)

log_cny_30 = restore_log_rate(mean_path_cny[30], w0_cny, trend_cny)
log_cny_180 = restore_log_rate(mean_path_cny[180], w0_cny, trend_cny)
log_cny_365 = restore_log_rate(mean_path_cny[365], w0_cny, trend_cny)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 取前 30/180/365 天
log_chf_test_30 = log_chf_test[:30]
log_chf_test_180 = log_chf_test[:180]
log_chf_test_365 = log_chf_test[:365]

log_cny_test_30 = log_cny_test[:30]
log_cny_test_180 = log_cny_test[:180]
log_cny_test_365 = log_cny_test[:365]

# 定义误差函数
def calculate_errors(true, pred):
    mse = mean_squared_error(true, pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(true, pred)
    return mse, rmse, mae

# 计算 CHF 误差
mse_chf_30, rmse_chf_30, mae_chf_30 = calculate_errors(log_chf_test_30, log_chf_30)
mse_chf_180, rmse_chf_180, mae_chf_180 = calculate_errors(log_chf_test_180, log_chf_180)
mse_chf_365, rmse_chf_365, mae_chf_365 = calculate_errors(log_chf_test_365, log_chf_365)

# 计算 CNY 误差
mse_cny_30, rmse_cny_30, mae_cny_30 = calculate_errors(log_cny_test_30, log_cny_30)
mse_cny_180, rmse_cny_180, mae_cny_180 = calculate_errors(log_cny_test_180, log_cny_180)
mse_cny_365, rmse_cny_365, mae_cny_365 = calculate_errors(log_cny_test_365, log_cny_365)

# 可视化函数
def plot_forecast_vs_actual(pred, true, title):
    plt.figure(figsize=(10, 4))
    plt.plot(true, label="Test (Actual)", linewidth=2)
    plt.plot(pred, label="Forecast (Mean Path)", linestyle="--")
    plt.title(title)
    plt.xlabel("Days")
    plt.ylabel("Log Exchange Rate")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 绘图：CHF
plot_forecast_vs_actual(log_chf_30, log_chf_test_30, "CHF Forecast vs Actual (30 days)")
plot_forecast_vs_actual(log_chf_180, log_chf_test_180, "CHF Forecast vs Actual (180 days)")
plot_forecast_vs_actual(log_chf_365, log_chf_test_365, "CHF Forecast vs Actual (365 days)")

# 绘图：CNY
plot_forecast_vs_actual(log_cny_30, log_cny_test_30, "CNY Forecast vs Actual (30 days)")
plot_forecast_vs_actual(log_cny_180, log_cny_test_180, "CNY Forecast vs Actual (180 days)")
plot_forecast_vs_actual(log_cny_365, log_cny_test_365, "CNY Forecast vs Actual (365 days)")

# 输出误差指标
import pandas as pd

results_df = pd.DataFrame({
    "Horizon": ["30 days", "180 days", "365 days"],
    "MSE_CHF": [mse_chf_30, mse_chf_180, mse_chf_365],
    "RMSE_CHF": [rmse_chf_30, rmse_chf_180, rmse_chf_365],
    "MAE_CHF": [mae_chf_30, mae_chf_180, mae_chf_365],
    "MSE_CNY": [mse_cny_30, mse_cny_180, mse_cny_365],
    "RMSE_CNY": [rmse_cny_30, rmse_cny_180, rmse_cny_365],
    "MAE_CNY": [mae_cny_30, mae_cny_180, mae_cny_365],
})

print(results_df)


```
