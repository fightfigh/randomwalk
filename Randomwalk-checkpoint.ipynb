```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
import pandas as pd

# 读取CSV文件
file_path = 'exchange_rates_2015_2024.csv'
df = pd.read_csv(file_path)

# 显示前几行数据
df.head()
```

```python
print(df.columns)
```

```python
df['observation_date'] = pd.to_datetime(df['observation_date'])
df.set_index('observation_date', inplace=True)

# 重命名列
df = df.rename(columns={'DEXSZUS': 'chf', 'DEXCHUS': 'cny'})

# 对 chf 和 cny 做对数变换
df_log = np.log(df[['chf', 'cny']])

# 绘制对数汇率时序图
df_log.plot(figsize=(12, 6), title='Log-Transformed Exchange Rates (CHF/USD & CNY/USD)')
plt.xlabel('Date')
plt.ylabel('Log Exchange Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
split_idx = int(len(df_log) * 0.8)
train = df_log.iloc[:split_idx].copy()
test = df_log.iloc[split_idx:].copy()

# 将训练集的 cny 和 chf 放大 100 倍
train_scaled = train * 100

# 绘制放大后的时序图
train_scaled.plot(figsize=(12, 6), title='Scaled Training Set (CNY & CHF) x 100')
plt.xlabel('Date')
plt.ylabel('Log Exchange Rate × 100')
plt.grid(True)
plt.tight_layout()
plt.show()

# ADF 单位根检验
print("ADF Test - CNY:")
print(adfuller(train_scaled['cny']))

print("\nADF Test - CHF:")
print(adfuller(train_scaled['chf']))

# 绘制 ACF 和 PACF
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
plot_acf(train_scaled['cny'], ax=axes[0, 0], title='ACF - CNY')
plot_pacf(train_scaled['cny'], ax=axes[0, 1], title='PACF - CNY', method='ywm')
plot_acf(train_scaled['chf'], ax=axes[1, 0], title='ACF - CHF')
plot_pacf(train_scaled['chf'], ax=axes[1, 1], title='PACF - CHF', method='ywm')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller, kpss
import matplotlib.pyplot as plt

# 一阶差分
train_diff = train_scaled.diff().dropna()

# 均值计算
cny_mean = train_diff['cny'].mean()
chf_mean = train_diff['chf'].mean()
print(f"CNY 差分均值: {cny_mean:.4f}")
print(f"CHF 差分均值: {chf_mean:.4f}")

# 分别绘制差分后的图
plt.figure(figsize=(12, 4))
plt.plot(train_diff['cny'], label='CNY First Difference')
plt.title('Differenced CNY (Log ×100)')
plt.xlabel('Date')
plt.ylabel('Difference')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(train_diff['chf'], label='CHF First Difference', color='orange')
plt.title('Differenced CHF (Log ×100)')
plt.xlabel('Date')
plt.ylabel('Difference')
plt.grid(True)
plt.tight_layout()
plt.show()

# ADF 检验
print("\nADF Test - Differenced CNY:")
print(adfuller(train_diff['cny']))

print("\nADF Test - Differenced CHF:")
print(adfuller(train_diff['chf']))

# KPSS 检验函数
def run_kpss(series, **kw):
    statistic, p_value, n_lags, crit_vals = kpss(series, **kw)
    return {
        'Test Statistic': statistic,
        'p-value': p_value,
        'Lags Used': n_lags,
        'Critical Values': crit_vals
    }

# KPSS 检验
print("\nKPSS Test - Differenced CNY:")
print(run_kpss(train_diff['cny'], regression='c'))

print("\nKPSS Test - Differenced CHF:")

```

```python
# 从之前计算得到的占比
# 示例：你应当在运行前先赋值 real_cny_counts 或 real_chf_counts，或用 resid_cny.round().value_counts(normalize=True)
# 这里用 cny_counts 举例：

# 获取 0 的原始概率
p0 = cny_counts.get(0, 0)

# 剩下的概率平分给 -1 和 1
remaining = 1 - p0
p_minus1 = remaining / 2
p_plus1 = remaining / 2

# 打印概率
probabilities = {
    -1: p_minus1,
     0: p0,
     1: p_plus1
}

print("调整后的状态转移概率：")
for state, prob in probabilities.items():
    print(f"状态 {state}: 概率 = {prob:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 模拟参数
n_simulations = 1000
days_list = [30, 90, 180]  # 模拟天数列表

# 状态与对应概率（从上一步获取）
states = [-1, 0, 1]

# 确保 cny_counts 和 chf_counts 已经定义
p0_cny = cny_counts.get(0, 0)
p_cny = {-1: (1 - p0_cny) / 2, 0: p0_cny, 1: (1 - p0_cny) / 2}

p0_chf = chf_counts.get(0, 0)
p_chf = {-1: (1 - p0_chf) / 2, 0: p0_chf, 1: (1 - p0_chf) / 2}

# 模拟函数
def simulate_mean_path(prob_dict, n_sim, n_days):
    probs = [prob_dict[s] for s in states]
    paths = np.random.choice(states, size=(n_sim, n_days), p=probs)
    return np.cumsum(paths, axis=1).mean(axis=0)

# 保存平均路径
mean_paths_cny = {}
mean_paths_chf = {}

# 模拟多个时间周期
for days in days_list:
    mean_paths_cny[days] = simulate_mean_path(p_cny, n_simulations, days)
    mean_paths_chf[days] = simulate_mean_path(p_chf, n_simulations, days)

# 打印 NumPy 数组形式结果
print("CNY 模拟平均路径（30天）：\n", mean_paths_cny[30])
print("\nCNY 模拟平均路径（90天）：\n", mean_paths_cny[90])
print("\nCNY 模拟平均路径（180天）：\n", mean_paths_cny[180])

print("\nCHF 模拟平均路径（30天）：\n", mean_paths_chf[30])
print("\nCHF 模拟平均路径（90天）：\n", mean_paths_chf[90])
print("\nCHF 模拟平均路径（180天）：\n", mean_paths_chf[180])
```

```python
# 还原函数（仅做累加，无需插入 w0）
def restore_level_series(diff_path):
    return np.cumsum(diff_path)

# 还原 CNY 模拟路径
restored_paths_cny = {days: restore_level_series(mean_paths_cny[days]) for days in [30, 90, 180]}
# 还原 CHF 模拟路径
restored_paths_chf = {days: restore_level_series(mean_paths_chf[days]) for days in [30, 90, 180]}
# 设置 NumPy 打印格式，避免科学计数法
np.set_printoptions(precision=6, suppress=True)

# 打印还原后的路径值
print("CNY 模拟还原路径（累加差分）：")
print("30天：", restored_paths_cny[30])
print("90天：", restored_paths_cny[90])
print("180天：", restored_paths_cny[180])

print("\nCHF 模拟还原路径（累加差分）：")
print("30天：", restored_paths_chf[30])
print("90天：", restored_paths_chf[90])
print("180天：", restored_paths_chf[180])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# 将 final path 除以 100
predicted_cny = {days: final_paths_cny[days] / 100 for days in [30, 90, 180]}
predicted_chf = {days: final_paths_chf[days] / 100 for days in [30, 90, 180]}

# 提取测试集实际值（前180天）
actual_cny = test['cny'].iloc[:180].reset_index(drop=True)
actual_chf = test['chf'].iloc[:180].reset_index(drop=True)

# 误差计算函数
def evaluate(pred, actual, days, label="CNY"):
    y_true = actual[:days]
    y_pred = pred[:days]
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    print(f"\n{label} - {days}天预测误差：")
    print(f"  MSE  = {mse:.6f}")
    print(f"  RMSE = {rmse:.6f}")
    print(f"  MAE  = {mae:.6f}")

# 计算并打印误差
for days in [30, 90, 180]:
    evaluate(predicted_cny[days], actual_cny, days, label="CNY")
    evaluate(predicted_chf[days], actual_chf, days, label="CHF")
```

```python
rmse_data = {
    "Days": [],
    "CNY_RMSE": [],
    "CHF_RMSE": []
}

for days in [30, 90, 180]:
    rmse_cny = 100*np.sqrt(mean_squared_error(actual_cny[:days], predicted_cny[days]))
    rmse_chf = 100*np.sqrt(mean_squared_error(actual_chf[:days], predicted_chf[days]))
    rmse_data["Days"].append(days)
    rmse_data["CNY_RMSE"].append(rmse_cny)
    rmse_data["CHF_RMSE"].append(rmse_chf)

# 构建 DataFrame 并打印
rmse_df = pd.DataFrame(rmse_data)
print(rmse_df)
```
