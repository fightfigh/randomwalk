```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# === Step 1: Load and prepare data ===
df = pd.read_csv("exchange_rates_log100.csv")
df['CHF'] = df['chf'] * 4
df['CNY'] = df['cny'] * 4
df = df[['observation_date', 'CHF', 'CNY']].dropna().reset_index(drop=True)

# Split into train/test
split_idx = int(len(df) * 0.8)
train_df = df.iloc[:split_idx]
test_df = df.iloc[split_idx:]

# Compute differenced residuals and trend a
train_diff_chf = train_df['CHF'].diff().dropna()
train_diff_cny = train_df['CNY'].diff().dropna()
a_chf = train_diff_chf.mean()
a_cny = train_diff_cny.mean()

# Round residuals and compute probability
rounded_chf = train_diff_chf.round()
rounded_cny = train_diff_cny.round()
chf_vals, chf_counts = np.unique(rounded_chf, return_counts=True)
cny_vals, cny_counts = np.unique(rounded_cny, return_counts=True)
chf_probs = dict(zip(chf_vals, chf_counts / chf_counts.sum()))
cny_probs = dict(zip(cny_vals, cny_counts / cny_counts.sum()))

# === Step 2: Symmetrize and normalize probabilities ===
def symmetrize_and_normalize(prob_dict, value_range):
    sym_probs = {}
    for k in value_range:
        if k < 0:
            avg = (prob_dict.get(k, 0) + prob_dict.get(-k, 0)) / 2
            sym_probs[k] = avg
            sym_probs[-k] = avg
        elif k == 0:
            sym_probs[0] = prob_dict.get(0, 0)
    total = sum(sym_probs.values())
    return {k: v / total for k, v in sym_probs.items()}

chf_sym = symmetrize_and_normalize(chf_probs, range(-4, 5))
cny_sym = symmetrize_and_normalize(cny_probs, range(-2, 3))

# === Step 3: Monte Carlo simulation ===
def sample_paths(prob_dict, a, start_val, path_len, n_sim=1000):
    values = np.array(list(prob_dict.keys()))
    probs = np.array(list(prob_dict.values()))
    paths = []
    for _ in range(n_sim):
        shocks = np.random.choice(values, size=path_len, p=probs)
        wt = np.cumsum(shocks + a) + start_val
        paths.append(wt)
    return np.array(paths)

# Parameters
path_lengths = [30, 180, 365]
n_sim = 1000
start_chf = train_df['CHF'].iloc[-1]
start_cny = train_df['CNY'].iloc[-1]
test_chf = test_df['CHF'].values
test_cny = test_df['CNY'].values

# === Step 4: Evaluate and plot ===
results = []

for name, sym_prob, a, start_val, test_val in [
    ("CHF", chf_sym, a_chf, start_chf, test_chf),
    ("CNY", cny_sym, a_cny, start_cny, test_cny)
]:
    for L in path_lengths:
        paths = sample_paths(sym_prob, a, start_val, L, n_sim)
        actual = test_val[:L]
        rmses = [np.sqrt(mean_squared_error(actual, path)) for path in paths]
        mean_rmse = np.mean(rmses)
        results.append({"Currency": name, "Days": L, "RMSE": mean_rmse})

        # Plot simulation paths
        plt.figure(figsize=(10, 4))
        for i in range(100):
            plt.plot(paths[i], alpha=0.3)
        plt.plot(actual, color='black', linewidth=2, label='Actual')
        plt.title(f'{name} - {L} Day Simulated Paths')
        plt.xlabel('Days')
        plt.ylabel('Value')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # Plot worst path
        worst_idx = np.argmax(rmses)
        worst_path = paths[worst_idx]
        plt.figure(figsize=(10, 4))
        plt.plot(worst_path, color='red', label='Worst Path')
        plt.plot(actual, color='black', label='Actual')
        plt.title(f'{name} - Worst Path (L={L})')
        plt.xlabel('Days')
        plt.ylabel('Value')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# === Step 5: RMSE summary table ===
rmse_df = pd.DataFrame(results)
print("\nRMSE Summary:")
print(rmse_df.pivot(index="Currency", columns="Days", values="RMSE").round(4))
```
