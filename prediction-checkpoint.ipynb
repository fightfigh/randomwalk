```python
import pandas as pd

# 1. 读取 CSV 文件
df = pd.read_csv("exchange_rates_2015_2024.csv")

# 2. 将日期列转换为 datetime 类型
df['observation_date'] = pd.to_datetime(df['observation_date'])

# 3. 将 observation_date 设置为索引（便于时间序列分析）
df.set_index('observation_date', inplace=True)

# 4. 查看数据结构和缺失值情况
print(df.info())

# 5. 显示前几行数据以确认加载正确
print(df.head())

```

```python
# 重命名列
df.rename(columns={'DEXSZUS': 'chf', 'DEXCHUS': 'cny'}, inplace=True)

# 检查是否修改成功
print(df.head())
```

```python
import numpy as np

# 设置 k 值
k = 100

# 对 chf 和 cny 取自然对数并乘以 k
df['log_chf_100'] = np.log(df['chf']) * k
df['log_cny_100'] = np.log(df['cny']) * k

# 查看结果
print(df[['chf', 'log_chf_100', 'cny', 'log_cny_100']].head())
```

```python
# 只划分对数化列


# 再划分为 80% 训练，20% 测试
split_index = int(len(df) * 0.8)
train_df = df.iloc[:split_index]
test_df = df.iloc[split_index:]
```

```python
c = train_df['log_chf_100'].iloc[-1]
d = train_df['log_cny_100'].iloc[-1]

print("c =", c)
print("d =", d)
```

```python
# 提取测试集中的 log_cny_100 和 log_chf_100 两列为两个数组
cny_test = test_df['log_cny_100'].to_numpy()
chf_test = test_df['log_chf_100'].to_numpy()

# 可选：查看前几个值
print("cny_test 前5项:", cny_test[:4])
print("chf_test 前5项:", chf_test[:4])
```

```python
chf_a = -0.0019
cny_d = 0.00423
n = len(chf_test)

steps = np.arange(1, n + 1)

# 去趋势处理：减去趋势项
chf_detrended = chf_test - chf_a * steps
cny_detrended = cny_test - cny_d * steps
```

```python
# 打印前10个 chf 残差值
print("chf_detrended 前10项：")
print(chf_detrended[:10])

# 打印前10个 cny 残差值
print("\ncny_detrended 前10项：")
print(cny_detrended[:10])
```

```python
# 从去趋势后的数组中减去训练集末值
chf_final = chf_detrended - c
cny_final = cny_detrended - d

# 打印前10项结果查看
print("chf_final 前10项：")
print(chf_final[:10])

print("\ncny_final 前10项：")
print(cny_final[:10])
```

```python
import numpy as np

# 给 chf_final 和 cny_final 前面加 0
chf_final_with_zero = np.insert(chf_final, 0, 0)
cny_final_with_zero = np.insert(cny_final, 0, 0)

# 打印前10项查看
print("chf_final_with_zero 前10项：")
print(chf_final_with_zero[:10])

print("\ncny_final_with_zero 前10项：")
print(cny_final_with_zero[:10])
```

```python
# 计算一阶差分（wt - wt-1 = et）
chf_diff = np.diff(chf_final_with_zero)
cny_diff = np.diff(cny_final_with_zero)

# 打印前10项结果查看
print("chf_diff 前10项：")
print(chf_diff[:10])

print("\ncny_diff 前10项：")
print(cny_diff[:10])
```

```python
# 假设已经有以下变量：
#   chf_diff, cny_diff    # 差分后的一阶残差数组
#   chf_test, cny_test    # 对数化并乘 100 后的原始测试数组
#   test_df               # 测试集 DataFrame

# 打印新数组（差分后）的长度
print("chf_diff 长度:", len(chf_diff))
print("cny_diff 长度:", len(cny_diff))

# 打印原始测试集的长度（观察日期行数）
print("测试集总长度:", len(test_df))

# （可选）或者直接打印原始测试数组的长度
print("chf_test 长度:", len(chf_test))
print("cny_test 长度:", len(cny_test))
```

```python
# 四舍五入取整
chf_rounded = np.round(chf_diff).astype(int)
cny_rounded = np.round(cny_diff).astype(int)

# 截断范围到 [-1, 1]
chf_discrete = np.clip(chf_rounded, -1, 1)
cny_discrete = np.clip(cny_rounded, -1, 1)

# 打印前10项查看
print("chf_discrete 前10项：", chf_discrete[:10])
print("cny_discrete 前10项：", cny_discrete[:10])
```

```python
import numpy as np

# 定义状态 [-1, 0, 1]
states = np.array([-1, 0, 1])
state_to_idx = {s: i for i, s in enumerate(states)}  # 映射状态值到索引

# 已有边际分布（你已有）
chf_probs = np.array([0.1114, 0.7772, 0.1114])
cny_probs = np.array([0.0266, 0.9468, 0.0266])

# 构造无记忆转换矩阵（你已有）
P_chf = np.outer(chf_probs, chf_probs)
P_cny = np.outer(cny_probs, cny_probs)

# 归一化每一行，确保作为 transition matrix 正确（随机选取下一状态时要求每行概率和为1）
P_chf = P_chf / P_chf.sum(axis=1, keepdims=True)
P_cny = P_cny / P_cny.sum(axis=1, keepdims=True)

# 示例原始状态序列（你也可以换成真实的）
# 例如你有 chf_discrete 和 cny_discrete，我们这里用随机序列来模拟（你可以替换）
N = len(chf_discrete)

# 初始化预测序列，长度加1，前面插入0
chf_et = np.zeros(N + 1, dtype=int)
cny_et = np.zeros(N + 1, dtype=int)

# 将原序列复制到预测序列的后部分
chf_et[1:] = chf_discrete
cny_et[1:] = cny_discrete

# 随机预测：基于当前状态，利用对应行的转移概率预测下一状态
for t in range(N):
    # --- CHF ---
    idx_chf = state_to_idx[chf_et[t]]
    chf_et[t + 1] = np.random.choice(states, p=P_chf[idx_chf])
    
    # --- CNY ---
    idx_cny = state_to_idx[cny_et[t]]
    cny_et[t + 1] = np.random.choice(states, p=P_cny[idx_cny])

# 打印前20个结果做验证
print("chf_et 长度:", len(chf_et), "前20项:", chf_et[:20])
print("cny_et 长度:", len(cny_et), "前20项:", cny_et[:20])



```

```python
import numpy as np


states = np.array([-1, 0, 1])
state_to_idx = {s: i for i, s in enumerate(states)}
N = len(chf_discrete)

# 1. 在每个离散序列前插入一个 0，得到长度 N+1 的预测数组
chf_et = np.zeros(N+1, dtype=int)
cny_et = np.zeros(N+1, dtype=int)

# 2. 滚动预测：用每个观测值 chf_discrete[t], cny_discrete[t] 来预测下一个 chf_et[t+1], cny_et[t+1]
for t in range(N):
    # CHF
    idx_chf = state_to_idx[chf_discrete[t]]
    chf_et[t+1] = states[np.argmax(P_chf[idx_chf])]
    # CNY
    idx_cny = state_to_idx[cny_discrete[t]]
    cny_et[t+1] = states[np.argmax(P_cny[idx_cny])]

# 3. 打印完整预测序列
print("chf_et:", chf_et)
print("cny_et:", cny_et)
```
